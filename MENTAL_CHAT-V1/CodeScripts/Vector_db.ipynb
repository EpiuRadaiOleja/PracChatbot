```python
from tqdm import tqdm
import pdfplumber
from langchain_core.documents import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
import chromadb
from sentence_transformers import SentenceTransformer
import pandas as pd
import os
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100, length_function=len)
```

```python
chunks = []
for pdf_file in tqdm(pdf_files, desc="Processing PDFs"):
    pdf_path = os.path.join(pdf_directory, pdf_file)
    pdf_name = pdf_file
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page_num, page in enumerate(pdf.pages, 1):
                page_text = page.extract_text()
                if page_text:
                    doc = Document(page_content=page_text, metadata={"pdf_name": pdf_name, "page_number": page_num})
                    chunks.extend(splitter.create_documents([page_text], metadatas=[doc.metadata]))
    except Exception as e:
        print(f"Failed to extract {pdf_file}: {e}")
```

```python
# Generate embeddings
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embeddings = []
for chunk in tqdm(chunks, desc="Encoding chunks"):
    embedding = model.encode(chunk.page_content, convert_to_tensor=False).tolist()
    embeddings.append(embedding)
```

```python
persist_directory = "vector_db"
client = chromadb.PersistentClient(path=persist_directory)
collection = client.get_or_create_collection(name="cervical_cancer")

for i, (chunk, embedding) in enumerate(tqdm(zip(chunks, embeddings), total=len(chunks), desc="Adding to ChromaDB")):
    collection.add(
        documents=[chunk.page_content],
        embeddings=[embedding],
        metadatas=[chunk.metadata],
        ids=[f"chunk_{i}"]
    )

print(f"Total chunks in Chroma DB: {collection.count()}")
```

```python
from langchain_core.documents import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
import chromadb
from tqdm import tqdm
import pandas as pd
import numpy as np
import pdfplumber
import os
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
persist_directory = "vector_db"
client = chromadb.PersistentClient(path=persist_directory)
collection = client.get_collection("cervical_cancer")
print(f"Total embeddings in collection: {collection.count()}")
```

```python
model_minilm = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

reranker_model_name = "cross-encoder/ms-marco-MiniLM-L-6-v2"
tokenizer = AutoTokenizer.from_pretrained(reranker_model_name)
reranker_model = AutoModelForSequenceClassification.from_pretrained(reranker_model_name)
device = "cuda" if torch.cuda.is_available() else "cpu"
reranker_model.to(device)
reranker_model.eval()

def retrieve_with_minilm(query, top_k=10):
    query_embedding = model_minilm.encode([query]).tolist()
    results = collection.query(query_embeddings=query_embedding, n_results=top_k)
    retrieved_ids = results["ids"][0]
    retrieved_texts = results["documents"][0]
    return retrieved_ids, retrieved_texts

def rerank(query, retrieved_ids, retrieved_texts):
    pairs = [(query, doc) for doc in retrieved_texts]
    inputs = tokenizer(pairs, padding=True, truncation=True, max_length=512, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = reranker_model(**inputs)
        scores = outputs.logits.squeeze(-1).cpu().tolist()

    scored = list(zip(scores, retrieved_ids, retrieved_texts))
    scored_sorted = sorted(scored, key=lambda x: x[0], reverse=True)
    reranked_ids = [item[1] for item in scored_sorted]
    reranked_texts = [item[2] for item in scored_sorted]
    return reranked_ids, reranked_texts

def handle_user_query(query, n_results=5):
    # Retrieve
    retrieved_ids, retrieved_texts = retrieve_with_minilm(query, top_k=10)

    # Rerank
    reranked_ids, reranked_texts = rerank(query, retrieved_ids, retrieved_texts)

    # Return top-n reranked results as response text
    response = "\n\n".join(reranked_texts[:n_results])
    return response
```

```python
from transformers import AutoProcessor, AutoModelForImageTextToText
import torch

# Load model and processor (even for text-only input)
model_id = "google/medgemma-4b-it"

model = AutoModelForImageTextToText.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float32,
    device_map="auto"
)
model.eval()

processor = AutoProcessor.from_pretrained(model_id)

def generate_medical_answer(prompt: str, context: str, max_new_tokens: int = 300) -> str:
    # Format in the way MedGemma expects (chat template)
    system_instruction = "You are a helpful and accurate medical assistant. Use the provided context to answer the question."

    messages = [
        {
            "role": "system",
            "content": [{"type": "text", "text": system_instruction}]
        },
        {
            "role": "user",
            "content": [{"type": "text", "text": f"Context:\n{context}\n\nQuestion:\n{prompt}"}]
        }
    ]

    # Prepare input
    inputs = processor.apply_chat_template(
        messages,
        add_generation_prompt=True,
        tokenize=True,
        return_dict=True,
        return_tensors="pt"
    ).to(model.device, dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float32)

    input_len = inputs["input_ids"].shape[-1]

    # Generate
    with torch.inference_mode():
        generation = model.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=False)
        generation = generation[0][input_len:]

    response = processor.decode(generation, skip_special_tokens=True)
    return response.strip()
```

```python
def handle_user_query(query, n_results=5):
    retrieved_ids, retrieved_texts = retrieve_with_minilm(query, top_k=10)
    reranked_ids, reranked_texts = rerank(query, retrieved_ids, retrieved_texts)
    context = "\n\n".join(reranked_texts[:n_results])
    response = generate_medical_answer(prompt=query, context=context)
    return response
```

# Llama Guard Inference

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
# Load LlamaGuard
model_id = "meta-llama/LlamaGuard-7b"
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.bfloat16 if torch.cuda.is_available() else torch.float32
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, torch_dtype=dtype, device_map=device)
model.eval()
```

```python
def moderate_with_template(chat: list) -> str:
    input_ids = tokenizer.apply_chat_template(chat, return_tensors="pt").to(model.device)
    with torch.no_grad():
        output = model.generate(input_ids=input_ids, max_new_tokens=100, pad_token_id=tokenizer.pad_token_id)
    prompt_len = input_ids.shape[-1]
    return tokenizer.decode(output[0][prompt_len:], skip_special_tokens=True).strip()
```

```python
print(moderate_with_template([
    {"role": "user", "content": "What are the side effects of chemotherapy?"}
]))
```
